```python
from reportlab.lib.pagesizes import letter
from reportlab.lib.units import inch
from reportlab.pdfgen import canvas
from reportlab.lib import colors

# Create a new PDF file
pdf = canvas.Canvas("example_report.pdf", pagesize=letter)

# Define some fonts
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont

styles = getSampleStyleSheet()
#pdfmetrics.registerFont(TTFont('Helvetica', 'Helvetica.ttf'))
#pdfmetrics.registerFont(TTFont('Helvetica-Bold', 'Helvetica-Bold.ttf'))
#pdfmetrics.registerFont(TTFont('Times-Roman', 'Times-Roman.ttf'))

# Define some colors
BLUE = colors.HexColor('#0074D9')
YELLOW = colors.HexColor('#FFDC00')
GRAY = colors.HexColor('#AAAAAA')

# Add a cover page
pdf.setFont('Helvetica-Bold', 36)
pdf.drawCentredString(4.25*inch, 9*inch, "Example PDF")
pdf.drawCentredString(4.25*inch, 8*inch, "Created with ReportLab")
pdf.showPage()

# Add a table of contents
pdf.setFont('Helvetica-Bold', 24)
pdf.drawCentredString(4.25*inch, 9*inch, "Table of Contents")
pdf.line(1*inch, 8.5*inch, 7.5*inch, 8.5*inch)
pdf.line(1*inch, 8.5*inch, 1*inch, 1*inch)
pdf.line(7.5*inch, 8.5*inch, 7.5*inch, 1*inch)
pdf.setFont('Helvetica', 12)
pdf.drawString(1.1*inch, 7.5*inch, "Chapter 1: Introduction")
pdf.drawString(1.1*inch, 6.5*inch, "Chapter 2: Methods")
pdf.drawString(1.1*inch, 5.5*inch, "Chapter 3: Results")
pdf.drawString(1.1*inch, 4.5*inch, "Chapter 4: Discussion")
pdf.drawString(1.1*inch, 3.5*inch, "Chapter 5: Conclusion")
pdf.showPage()

# Add chapter 1
pdf.setFont('Helvetica-Bold', 24)
pdf.drawCentredString(4.25*inch, 9*inch, "Chapter 1: Introduction")
pdf.setFont('Helvetica', 12)
pdf.drawString(1*inch, 8*inch, "test text1")
pdf.drawString(1*inch, 7.5*inch, "test text1")
pdf.drawString(1*inch, 7*inch, "test text1")
pdf.drawString(1*inch, 6.5*inch, "test text1")
pdf.showPage()

# Add chapter 2
pdf.setFont('Helvetica-Bold', 24)
pdf.drawCentredString(4.25*inch, 9*inch, "Chapter 2: Methods")
pdf.setFont('Helvetica', 12)
pdf.drawString(1*inch, 8*inch, "test text1")
pdf.drawString(1*inch, 7.5*inch, "test text1")

# Save the PDF
pdf.save()
```
